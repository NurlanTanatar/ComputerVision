## BGR/RGB examples

```python
car = cv2.imread('assets/car.jpg')
car.shape
```

```python
plt.imshow(car)
plt.show()
```

### Changing channels order convention

```python
carRGB = cv2.cvtColor(car, cv2.COLOR_BGR2RGB)
plt.imshow(carRGB)
plt.show()
```

### Let's see different channels

```python
plt.imshow(car[:, :, 2], cmap='gray') # red
plt.show()
```

```python
plt.imshow(car[:, :, 0], cmap='gray') # blue
plt.show()
```

```python
plt.imshow(car[:, :, 1], cmap='gray') # green
plt.show()
```

```python
gray_manual = car[..., 0] * 0.114 + car[..., 1] * 0.587 + car[..., 2] * 0.299
print(type(gray_manual))
print(type(gray_manual[0, 0]))
```

```python
plt.imshow(gray_manual, cmap='gray')
plt.show()
```

```python
gray_manual = gray_manual.astype(np.uint8)
```

### OpenCV gray

```python
gray_opencv = cv2.cvtColor(car, cv2.COLOR_BGR2GRAY)
print(type(gray_opencv))
print(type(gray_opencv[0, 0]))
plt.imshow(gray_opencv, cmap='gray')
plt.show()
```

### Make white square image with black box inside

```python
img2 = np.full((500, 500), 255, dtype = np.uint8)
img2[200:300, 200:300] = 0
plt.imshow(img2, cmap='gray')
plt.show()
```

### Filtering

```python
kernel = np.ones((5, 5), dtype = np.float64) / 25
kernel
manual_filter = cv2.filter2D(gray_opencv, - 1, kernel)
plt.imshow(manual_filter, cmap='gray')
plt.show()
```

```python
opencv_filter = cv2.blur(gray_opencv, (5,5))
plt.imshow(opencv_filter, cmap='gray')
plt.show()
```

```python
opencv_filter = cv2.GaussianBlur(gray_opencv, (5,5), 1)
plt.imshow(opencv_filter, cmap='gray')
plt.show()
```
